Before going through this tutorial, you should learn how to run basic density functional theory (DFT) calculations: complete modules 0 and 1 of http://wiki.max-centre.eu/index.php/LabQSM.

In this tutorial, we will calculate the band structure of bulk silicon using Quantum Espresso and project it on atomic orbitals. We will then plot the result in two ways: as a projected density of states and as a orbital-resolved band structure, also called "fatbands".

````python
# Libraries
import numpy as np
import matplotlib.pyplot as plt
from ase.io import read, write
from ase.visualize import view
from bs import bandPlotSimpleWithFatbands  # This is a homemade bandstructure plotting function
````

We start by running a simple self-consistent field (SCF) calculation for silicon using `pw.x`.
``` 
&control
    calculation = 'scf'
    prefix ='Si',
    pseudo_dir = '.',
    outdir = 'tmp/'
/
&system
    ibrav =  2,
    celldm(1) = 10.20,
    nat =  2,
    ntyp = 1,
    ecutwfc = 18.0,
    occupations ='smearing',
    smearing ='methfessel-paxton',
    degauss = 0.05,
/
&electrons
    conv_thr =  1.0d-8
/
ATOMIC_SPECIES
 Si  28.086  Si.pz-vbc.UPF
ATOMIC_POSITIONS alat
 Si 0.00 0.00 0.00
 Si 0.25 0.25 0.25
K_POINTS {automatic}
4 4 4 0 0 0
```
We now want to calculate the Kohn-Sham eigenvalues and eigenvectors along a particular path in the Brillouin zone, using the ground state density calculated in this SCF calculation. To choose this path, we start by plotting the Brillouin zone and its high-symmetry points.

````python
inputFile = read("Si.scf.in",format="espresso-in") # Read the input file
lattice = inputFile.cell.get_bravais_lattice()     # Find the corresponding Bravais lattice
print(lattice.description())                       # Print the description of the lattice
lattice.plot_bz(show=True)                         # Plot the Brillouin zone
plt.show()
````

We then choose the path G-X-U-K-G. We now can run a non-self-consistent field (NSCF) calculation using `pw.x` along this path.

```
&control
    calculation='bands'
    prefix='Si'
    pseudo_dir = '.',
    outdir='tmp/',
/
&system
    ibrav =  2,
    celldm(1) = 10.20,
    nat =  2,
    ntyp = 1,
    ecutwfc = 18.0,
    nbnd = 8,
/
&electrons
/
ATOMIC_SPECIES
 Si  28.086  Si.pz-vbc.UPF
ATOMIC_POSITIONS alat
 Si 0.00 0.00 0.00
 Si 0.25 0.25 0.25
K_POINTS {crystal_b}
5
  0.00000  0.00000  0.00000    20
  0.00000  0.50000  0.50000    20
  0.25000  0.62500  0.62500    20
  0.37500  0.7500   0.37500    20
  0.00000  0.00000  0.00000    0
```

Finally, we use the post-processing tool `bands.x` to post-process the output of this calculation.

```
&bands
prefix  = "Si",
outdir = "tmp/"
/
```

Note that this calculation sometimes crashes if there is no new line at the end of the input file. This produces an output file `bands.out.gnu` that we can plot by running the command `bandPlotSimpleWithFatbands`. Its first three arguments are paths to the SCF, bands, and band plotting output files.

````python
symmetryPointsNames = ['$\Gamma$','$X$','$U$','$K$','$\Gamma$']
bandPlotSimpleWithFatbands('Si.scf.out','Si.bands.out','bands.out.gnu', symmetryPointsNames,plt.subplot(1,1,1))
plt.show()
````

We now want to project the Kohn-Sham wavefunctions on atomic orbitals, to identify the contribution of each orbital to the density of states (DOS), using `projwfc.x`. The result of this calculation is the projected density of states (PDOS).
```
&projwfc
  prefix='Si',
  outdir = 'tmp/'
  degauss = 0.001,             # Broadening
  DeltaE = 0.001,              # Energy step
  Emin = -10,                  # Energy range
  Emax = 20,
  filproj = 'bands.out.proj'   # Output file
/
```
The corresponding output file `Si.projwfc.out` contains the list of atomic states used for projection. $s$ orbitals are numbered 1 and 5, while $p$ orbitals are numbered 2, 3, 4, 6, 7 and 8. Note that there are only 3 $p$ orbitals per atoms, since the calculation is not spin-polarised. There are four other output files such as `Si.pdos_atm#1(Si)_wfc#1(s)` that contain the PDOS as a function of energy, for each atom and each orbital.

Finally, we want to plot how the weight over some orbitals varies in the Brillouin zone. For this, we run `plotband.x`. It is run interactively, i.e. without using an input file. Here are the parameters used to project each Kohn-Sham wavefunction on the $p$ orbitals:
```
bands.out 		! Input file for bands
2 3 4 6 7 8		! Wave-functions on which to project
1 100			! First and last k-point
Si.p			! Output file for xmgrace
Si.ps 			! Output postscript file
6.5634			! Fermi energy
0.01, 6.5634	! Energy step, zero of the plot
plot.gnu		! Plotting script for gnuplot
```
It is then possible to plot the result exactly like a normal band structure, but using the output file for xmgrace produced by `plotband.x`.

````python
bandPlotSimpleWithFatbands('Si.scf.out','Si.bands.out','Si.p', symmetryPointsNames,plt.subplot(1,1,1))
plt.show()
````
